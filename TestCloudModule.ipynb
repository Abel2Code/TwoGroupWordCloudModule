```python
from CloudModule import generateTwoGroupWordCloud, DEFAULT_COLORS
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
text1 = "The quick brown fox jumps over the lazy #dog."
text2 = "The fox chased the #chicken around the dog house!"
```

# Generate Word Cloud

```python
wc = generateTwoGroupWordCloud(text1, text2)
```

# Plot

```python
# Create Figure
plt.figure(figsize=(14,10))
plt.imshow(wc, interpolation="bilinear")

# Add to Figure
plt.title(f"Word Cloud")
plt.axis("off")

# Add Legend
group_1_patch = mpatches.Patch(color=DEFAULT_COLORS[-1], label='Group 1')
group_2_patch = mpatches.Patch(color=DEFAULT_COLORS[0], label='Group 2')
plt.legend(handles=[group_1_patch, group_2_patch])

# Show Figure
plt.show()
```
